```python
#ประกาศเพื่อเรียกใช้งาน 
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
#Lab01
sin_wave=thinkdsp.SinSignal(freq=7001,amp=1,offset=0)
sin_wave.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#Lab02
#Trianglesignal(freq)
#duration = signal.period*n 
signal = thinkdsp.TriangleSignal(1)
duration = signal.period*1
segment = signal.make_wave(duration,framerate=10000)
segment.plot()

snwave=signal.make_wave(duration,framerate=10000)
snwave.apodize()
snwave.make_audio()

spectrum=snwave.make_spectrum()
spectrum.plot()
```

```python
signal = thinkdsp.SquareSignal(1)
duration = signal.period*1
segment = signal.make_wave(duration,framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05,1.05],legend=False)
suwave=signal.make_wave(duration,framerate=10000)
suwave.apodize()
suwave.make_audio()
```
